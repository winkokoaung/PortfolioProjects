```python
data = pd.read_csv ('countries.csv')
```

```python
data.head()
```

```python
# Compare the population growth in Myanmar and Thailand
```

```python
MM = data[data.country == "Myanmar"]
TH = data[data.country == "Thailand"]
KH = data[data.country == "Cambodia"]
VN = data[data.country == "Vietnam"]
CN = data[data.country == "China"]
```

```python
plt.plot(MM.year, MM.population / 10**6)
plt.plot(TH.year, TH.population / 10**6)
plt.plot(KH.year, KH.population / 10**6)
plt.plot(VN.year, VN.population / 10**6)
plt.legend (['Myanmar', "Thailand", "Cambodia", "VietNam", "China"])
plt.xlabel ('year')
plt.ylabel ('population')
plt.show()
```

```python
plt.plot(MM.year, MM.population / MM.population.iloc[0] * 100)
plt.plot(TH.year, TH.population / TH.population.iloc[0] * 100)
plt.plot(KH.year, KH.population / KH.population.iloc[0] * 100)
plt.plot(VN.year, VN.population / VN.population.iloc[0] * 100)
plt.plot(CN.year, CN.population / CN.population.iloc[0] * 100)
plt.legend (['Myanmar', "Thailand", "Cambodia", "VietNam", "China"])
plt.xlabel ('year')
plt.ylabel ('population growth (first year = 100)')
plt.show()
```
